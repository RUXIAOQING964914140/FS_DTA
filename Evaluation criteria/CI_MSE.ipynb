```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
import sklearn
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
%matplotlib inline
```

```python
from lifelines.utils import concordance_index
b=pd.read_csv("test_affi.csv")
df = pd.DataFrame(b)
m = df['Y']
n = df['P']
r2 = sklearn.metrics.r2_score(m,n)
print(r2)
print(f'Concordance index: {concordance_index(m, n)}')
print(f"均方误差(MSE)：{mean_squared_error(m, n)}")
```
